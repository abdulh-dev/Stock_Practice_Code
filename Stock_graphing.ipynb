```python
#Now we are gonna go more indepth with plotly
import plotly.offline as pyo
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf 
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#pyo.init_notebook_mode(connected=True) used previously to link plotly to jupyternotebook
```

```python
end = dt.datetime.now()
start = dt.datetime(2020,1,1)
print(start,'\n' ,end)
```

```python
#Selecting stocks

yf.pdr_override()#don't know just use to deal with str and indices problem/ may be use a different library

# pulling stocks
df = pdr.get_data_yahoo('NFLX', start, end)
df.head()
```

```python
#Create moving average terms
#We are going to use .rolling function from pandas 
df['MA50']= df['Close'].rolling(window=50, min_periods=0).mean()
#.rolling(window=10).mean() takes 10 preceeding values to make a mean but this makes it so that the first 10 entries are Nan
#to deal with this we  min_periods=0 which makes it so that it will give us a mean with whatever data it can get

df['MA200']= df['Close'].rolling(window=200, min_periods=0).mean()
df['MA50'].head(20)
```

```python
#Creating pyplot subplot format before placing data in the fig
fig = make_subplots(rows=2, cols=1, shared_xaxes =True,
                    vertical_spacing=0.1, subplot_titles=('NFLX', 'Volume'),
                    row_width=[0.2,0.7])
```

```python
#Adding open, high, low, close and volume to fig
fig.add_trace(go.Candlestick(x=df.index, open=df['Open'], high=df['High'], low=df['Low'], close=df['Close']), row=1, col=1)
#Volume is not complete yet
```

```python
#Adding moving average terms to fig
fig.add_trace(go.Scatter(x=df.index, y=df['MA50'], marker_color='grey',), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df['MA200'], marker_color='lightgrey',), row=1, col=1)
#Note: one scatterplot mode the default is a line graph
```

```python
#adding volume bar chart
fig.add_trace(go.Bar(x=df.index, y=df['Volume'], marker_color='red', showlegend=False), row=2, col=1)
```

```python
#upadting labels and chart format
fig.update_layout(
    title = 'NFLX price chart',
    xaxis_tickfont_size =12,
    yaxis = dict(
        title = 'Price ($/share)',
        titlefont_size=14,
        tickfont_size=12
    ),
    autosize=False,
    width=800,
    height=500,
    margin=dict(l=50,r=50,b=100,t=100,pad=5),
    paper_bgcolor = 'LightSteelBlue'
)
```

```python
#Removing rangeslider
fig.update(layout_xaxis_rangeslider_visible=False)
fig.show()
```
